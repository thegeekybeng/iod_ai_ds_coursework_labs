````python
# Best Workaround solution that I can deploy to get the exact same output

# sample points
X = [0, 5, 10, 15, 20]
Y = [0, 7, 10, 13, 20]

# Function to calculate best fit line parameters
def best_fit(X, Y):
    N = len(X)  # Number of data points

    # Compute sums needed for the slope formula
    sum_x = sum(X)
    sum_y = sum(Y)
    sum_x2 = sum(x**2 for x in X)
    sum_xy = sum(x*y for x, y in zip(X, Y))

    m = (N * sum_xy - sum_x * sum_y) / (N * sum_x2 - sum_x**2)
    b = (sum_y - m * sum_x) / N

    print('best fit line:\ny = {:.2f} + {:.2f}x'.format(b, m))

    return b, m

# solution
b, m = best_fit(X, Y)

# plot points and fit line
# ANSWER
plt.scatter(X, Y);
yfit = [b + m * xi for xi in X]
plt.plot(X, yfit);
plt.show()
````

````python
fig = plt.figure()
ax1 = fig.add_subplot(211)
np.random.seed(0)
x = stats.loggamma.rvs(5, size=500) + 5
# x follows a loggamma (i.e. non-bell-shaped) distribution - hence its probplot will not be linear at either end
# A Box-Cox transformation of x can achieve a more bell-shaped-looking distribution
# ANSWER
# Generate probability plot (QQ plot) against normal distribution
stats.probplot(x, dist="norm", plot=ax1)
ax1.set_title("Probplot against normal distribution")

# Add second subplot (Probplot after Box-Cox transformation)
ax2 = fig.add_subplot(212)

# Apply Box-Cox transformation (to normalize x)
x_transformed, _ = stats.boxcox(x)

# Generate probability plot for transformed data
stats.probplot(x_transformed, dist="norm", plot=ax2)
ax2.set_title("Probplot after Box-Cox transformation")
# Visualise probability plot
plt.show()
````
